```python
# Replace 'Error' in "Annual Salary" with NaN

df['Annual Salary'] = pd.to_numeric(df['Annual Salary'], errors = 'coerce')
```

```python
# replace 'UNKNOWN' in 'City' with NaN
df['City'].replace('UNKNOWN', np.nan, inplace = True)
```

```python
# Fill Missing 'City' with Mode
city_mode = df['City'].mode()[0]
df['City'].fillna(city_mode, inplace=True)
```

```python
# to find if data is normally distribute or not
df["Annual Salary"].skew()
```

```python
# replace empty annual salary with median
AS_median = df["Annual Salary"].median()
df["Annual Salary"].fillna(AS_median, inplace=True)
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors = 'coerce')

# filter the DataFrame
print("\nTransactions with Age > 50:")
older_age = df[df["Age"] > 50]
print(older_age)
```

```python
# selection based on conditions
filtered_conditions = df[(df["Age"] > 50) | (df["City"] == "Chicago")]
print(filtered_conditions)
```

```python
# filter transactions for specific items (e.g., "Sr. Analyst " or "Director")
selected_items = df[df["Job Title"].isin(["Sr. Analyst ", "Director"])]
print(selected_items)
```

```python
# CREATING NEW COLUMN OF GET BONUS
df.loc[(df["Bonus %"] == 0), "Get Bonus"] = "no bonus"
df.loc[(df["Bonus %"] > 0), "Get Bonus"] = "bonus"
print(df.head(10))
```

```python
# Group by Country and sum Annual Salary
print("\nAnnual Salary by Country:")
grouped_data = df.groupby("Country")["Annual Salary"].sum()
print(grouped_data)
```

```python
# Aggregating Annual Salary by Country (Mean & Count)
salary_grouped = df.groupby("Country")["Annual Salary"].agg(["mean", "count"])
print(salary_grouped)

# Aggregating Bonus % by Department (Mean & Count)
bonus_grouped = df.groupby("Department")["Bonus %"].agg(["mean", "count"])
print(bonus_grouped)
```

# MATPLOTLIB AND SEABORN CHARTS

```python
import matplotlib.pyplot as plt
import seaborn as sns

pd.crosstab(df['Department'], df['Gender']).plot(kind='bar', stacked=True, figsize=(10,5),  colormap='viridis')
plt.title("Stacked Bar Chart: Department vs. Gender Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Gender")
plt.show()
```

Key Insights:

The IT department has the highest number of employees, with a higher male presence.

Engineering also has more males than females, but the difference is smaller.
Finance, Human Resources, and Sales have a relatively more balanced distribution.

Human Resources appears to have more females than males.

```python
sns.histplot(data=df, x="Annual Salary", kde=True, hue="Business Unit", bins=20)
plt.title("Distribution Plot of Annual Salary by business unit")
plt.show()
```

**Key Insights**

Most employees have salaries between $50,000 and $150,000, with the highest concentration around $75,000 - $100,000.

Manufacturing (orange) has the highest frequency in the lower salary range.

Corporate (red) and Specialty Products (green) show a more even distribution across salary ranges, including higher salaries.

The density curves (KDE) suggest that higher salaries are less common, but Corporate and Specialty Products have some employees earning over $200,000.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Select relevant numerical columns
num_cols = ["Annual Salary", "Bonus %", "Age"]  # Adjust based on dataset

# Compute correlation matrix
corr_matrix = df[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

Key Insights
Annual Salary & Bonus % (0.92)

Strong correlation: Employees with higher salaries also tend to receive higher bonuses.
Likely due to performance-based or percentage-based bonus structures.

Age vs. Salary (-0.03) & Age vs. Bonus % (-0.02)
Very weak correlation, meaning age is not a strong predictor of salary or bonus percentage.
Salary might depend more on experience, job role, or business unit rather than just age.

```python
plt.figure(figsize=(10,5))
sns.barplot(x="Department", y="Bonus %", data=df, estimator=np.mean, ci=None, palette="coolwarm")

plt.title("Average Bonus % by Department", fontsize=14)
plt.xlabel("Department", fontsize=12)
plt.ylabel("Average Bonus (%)", fontsize=12)
plt.xticks(rotation=45)

plt.show()
```

- Accounting & Human Resources also have relatively high bonuses (~11-12%).

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Business Unit", y="Annual Salary", data=df, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Annual Salary Distribution Across Business Units")
plt.xlabel("Business Unit")
plt.ylabel("Annual Salary")
plt.show()

```

Key Insights:

All business units have a similar median salary (middle line inside the boxes).

Manufacturing has more employees with lower salaries, as seen by its lower quartile being lower than other units.

Research & Development and Corporate have the highest salary ranges, extending above $250,000.

Manufacturing has more outliers, suggesting a few employees earn significantly higher salaries.

```python
# Filter Annual Salary for Manufacturing Business Unit
manufacturing_salaries = df[df["Business Unit"] == "Manufacturing"]["Annual Salary"]

# Compute IQR
Q1 = manufacturing_salaries.quantile(0.25)
Q3 = manufacturing_salaries.quantile(0.75)
IQR = Q3 - Q1

# Define outlier thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df_cleaned = df[~((df["Business Unit"] == "Manufacturing") &
                   ((df["Annual Salary"] < lower_bound) | (df["Annual Salary"] > upper_bound)))]
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Business Unit", y="Annual Salary", data=df_cleaned, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Annual Salary Distribution After Removing Outliers in Manufacturing")
plt.xlabel("Business Unit")
plt.ylabel("Annual Salary")
plt.show()
```
